```python
import numpy as np
import matplotlib.pyplot as plt
import pandas
import numpy as np
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

def getMean(list) :
    lengthOfList = len(list)
    sumOfElement = 0
    for i in range((lengthOfList) - 2):
        sumOfElement += float(list[i+2])
    return sumOfElement/(len(list) - 2)



f = open('./douglas_data.csv', 'r')
csvReader = csv.reader(f)

list1 = []
list2 = []
list3 = []
list4 = []
list5 = []
list6 = []
list7 = []
list8 = []
for row in csvReader:
    list1.append(row[1])
    list2.append(row[2])
    list3.append(row[3])
    list4.append(row[4])
    list5.append(row[5])
    list6.append(row[6])
    list7.append(row[7])
    list8.append(row[8])

print("The mean of "+ list1[0] + " is " + str(getMean(list1)))
print("The mean of "+ list2[0] + " is " + str(getMean(list2)))
print("The mean of "+ list3[0] + " is " + str(getMean(list3)))
print("The mean of "+ list4[0] + " is " + str(getMean(list4)))
print("The mean of "+ list5[0] + " is " + str(getMean(list5)))
print("The mean of "+ list6[0] + " is " + str(getMean(list6)))
print("The mean of "+ list7[0] + " is " + str(getMean(list7)))
print("The mean of "+ list8[0] + " is " + str(getMean(list8)))

set = { list1[0] ,list2[0] 
        ,list3[0] ,list4[0] 
        ,list5[0] ,list6[0] 
        ,list7[0] ,list8[0]
       }

vari =  { getMean(list1), getMean(list2)
        , getMean(list3), getMean(list4)
        , getMean(list5), getMean(list6)
        , getMean(list7), getMean(list8)
        }

# Number of groups
num_groups = len(vari)

# Create the figure
fig, ax = plt.subplots()

# Define the X axis
indices = np.arange(num_groups)

# Width and opacity of histogram bars
bar_width = 0.4
opacity = 0.6

# Plot the values

hist_people = plt.bar(indices , vari, bar_width,
        alpha=opacity, color='b', label='Standard')

plt.xlabel('Time')
plt.ylabel('# Of People Got Off')
plt.title('# Of People By Time')
plt.xticks(indices + bar_width, set)
plt.ylim([0, 10000])
plt.legend()
plt.tight_layout()
fig.autofmt_xdate()
plt.show()
f.close
```
