```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy.random as rnd
```

```python
n = 10
p = 0.3
k = np.arange(0, 21)
binomial = stats.binom.pmf(k, n, p)
binomial
```

```python
plt.plot(k, binomial, '-o')
plt.title('Binomial: n=%i , p=%.2f' % (n,p), fontsize=15)
plt.xlabel('Numer of successes')
plt.ylabel('Probability of successes', fontsize=15)
plt.show()
```

```python
binom_sim = data = stats.binom.rvs(n = 10, p = 0.3, size = 10000)
print ('Mean: %g' % np.mean(binom_sim))
print ('SD: %g' % np.std(binom_sim, ddof=1))
plt.hist(binom_sim, bins = 10, normed = True) #normed
plt.xlabel('x')
plt.ylabel('density')
plt.show()
```

```python
p = 0.35
k = np.arange(0, 21)
geom = stats.geom.pmf(k, p)
geom
```

```python
plt.plot(k, geom, '-o')
plt.title('Geometric: $p$ = %i' % p)
plt.xlabel('Numer of trials')
plt.ylabel('Probability of trials', fontsize=15)
plt.show()
```

```python
data = stats.geom.rvs(p=0.35, size = 1000)
print('Mean: %g' % np.mean(data))
print('SD: %g' % np.std(data, ddof=1))

plt.figure()
plt.hist(data, bins = 19, normed = True)
plt.xlim(0, 20)
plt.xlabel('Number of trials')
plt.title('Simulating Geometric Random Variables')
plt.show()
```

```python
rate = 2
n = np.arange(0, 10)
y = stats.poisson.pmf(n, rate)
y
```

```python
plt.plot(n, y, 'o-')
plt.title('Poisson: $\lambda$ = %i' % rate)
plt.xlabel('Number of accidents')
plt.ylabel('Probability of numer of accidents')
plt.show()
```

```python
data = stats.poisson.rvs(mu = 2, loc = 0, size = 1000)
print('Mean: %g' % np.mean(data))
print('SD: %g' % np.std(data, ddof=1))

plt.figure()
plt.hist(data, bins = 9, normed = True)
plt.xlim(0, 10)
plt.xlabel('Number of accidents')
plt.title('Simulating Poisson Random Variables')
plt.show()
```

```python
mu = 0 # mean
sigma = 1 # standard deviation

x = np.arange(-5, 5, 0.1)
y = stats.norm.pdf(x, 0, 1)

plt.plot(x, y)
plt.title('Normal: $\mu$=%.1f, $\sigma^2$=%.1f' % (mu, sigma))
plt.xlabel('x')
plt.ylabel('Probability density')
plt.show()
```

```python
lambd = 0.5
x = np.arange(0, 15, 0.1)
y = lambd * np.exp(-lambd * x)   # could also use stats.expon.pdf

plt.plot(x, y)
plt.title('Exponential: $\lambda$ =%.2f' % lambd)
plt.xlabel('x')
plt.ylabel('Probability density')
plt.show()
```

```python
data = stats.expon.rvs(scale = 2, size = 1000)

print ('Mean: %g' % np.mean(data))
print ('SD: %g' % np.std(data, ddof=1))

plt.figure()
plt.hist(data, bins = 20, normed = True)
plt.xlim(0, 15)
plt.title('Simulating Exponential Random Variables')
plt.show()
```
