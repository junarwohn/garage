```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as rnd
from pandas import read_csv
import pandas
from scipy import stats
%matplotlib inline

def getAvg(arr) :
    totalN = len(arr)
    sumOfElement = 0
    for element in arr :
        sumOfElement = sumOfElement + float(element)
    return sumOfElement/totalN

def getVar(arr, avg) :
    totalN = len(arr)
    sumOfSqure = 0
    for element in arr :
        sumOfSqure = sumOfSqure + (float(element) - avg) * (float(element) - avg)
    return sumOfSqure/(totalN - 1)

def getMin(arr) :
    minEle = arr[0]
    for element in arr :
        if (minEle > element) :
            minEle = element
    return minEle


def getMax(arr) :
    maxEle = arr[0]
    for element in arr :
        if (maxEle < element) :
            maxEle = element
    return maxEle
    
def z_test(x, avg, var, arr) :
    ele = len(arr)
    return ((x - avg)/ math.sqrt(var)) * math.sqrt(ele)

def t_test(x, avg, var, arr) :
    ele = len(arr)
    return ((x - avg)/ math.sqrt(var)) * math.sqrt(ele)

def drawBoxPlot(arr) :
    arr_ = []
    for i in arr:
        arr_.append(float(i))
    a = plt.boxplot(arr_)

def drawHistogram(arr) :
    fig = plt.figure()
    arr_ = []
    for i in arr :
        arr_.append(float(i))
    maxVal = getMax(arr)
    minVal = getMin(arr)
    maxRough = int(float(maxVal)/10) * 10
    minRough = int(float(minVal)/10) * 10
    rangeLen = int((maxRough - minRough)/10 + 1)
    boxArr = []
    for i in range(rangeLen) :
        boxArr.append(0)
    for element in arr_ :
        index = int((element - float(minVal))/10)
        boxArr[index] = boxArr[index] + 1
    x_set = []
    for i in range(rangeLen) :
        x_set.append(str(minRough + 10*(i)) + "-" + str(minRough + 10*(i+1)))
    indices = np.arange(len(x_set))
    bar_width = 0.4
    opacity = 0.6
    histGram = plt.bar(indices, boxArr, bar_width,
                      alpha=opacity, color='b', label="Strength")
    plt.xlabel("BendingStrength")
    plt.ylabel("# of Wooden Beam")
    plt.title("Distribution of Wooden Beam")
    plt.xticks(indices + bar_width, x_set)
    plt.ylim([0,20])
    plt.legend()
    plt.tight_layout()
    fig.autofmt_xdate()
    plt.show()
    
def showPvalue(z) : 
    if z < 0 :
        print("P value is " + str(1 - stats.chi2.pdf(-z, 1)))
    else :
        print("P value is " + str(stats.chi2.pdf(z, 1)))
```
