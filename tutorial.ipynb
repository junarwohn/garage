```python
# array statistic

# sum
arr_a = [1,3,5,7,9]
sum_of_arr = sum(arr_a)
print("sum of array : {}".format(sum_of_arr))

# mean
length_of_arr = len(arr_a)
mean_of_arr = sum_of_arr / length_of_arr
print("mean of array : {}".format(mean_of_arr))

# squre

var_a = 10
squared_var_a = var_a * var_a
# or
squared_var_a = var_a ** 2
print("square of var_a : {} -> squared_var_a : {}".format(var_a, squared_var_a))

# Mean Squared Error
MSE = 0
for e in arr_a:
    MSE += (e - mean_of_arr)**2

MSE /= length_of_arr
print("mean squared error : {}".format(MSE))
```

```python
# numpy array generation
# make every function we need is so painfullll.
# We use mathmetical libray numpy, tensorflow, pytorch..etc 

import numpy as np
# becuase numpy is 5 letters and np is 2 letters....
arr_a = [1,2,3,4,5]
arr_a_np = np.array(arr_a)
print("numpy array arr_a_np : {}".format(arr_a_np))
```

```python
# numpy statistics
# actually, numpy treat array more like a Vector, Matrix, Tensor...etc, not like list of element as above.
# now we have wings to fly
arr_a = [1,3,5,7,9]
arr_a_np = np.array(arr_a)

# sum
sum_arr_a = np.sum(arr_a_np)

# mean
mean_arr_a = np.mean(arr_a_np)

# substract value A from array B! (B - A)
val = 3
print("substract val : {} from arr_a_np : {}".format(val, arr_a_np))
print("arr_a_np - val : {}".format(arr_a_np - val))

# Mean Squared Error
# As told before, numpy treat 1D array as vector,
# so we can get the sum of squared element by squaring vector

MSE = np.mean(np.square(arr_a_np - mean_arr_a))
print("Mean Squared Error : {}".format(MSE))
```
